```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading the dataset Iris into Python environment

```python
data=pd.read_csv('iris.csv')
```

# Displaying the columns of the dataset

```python
data.columns
```

# Mean of all columns in dataset

```python
iris=pd.DataFrame(data,columns=['SL','SW','PL','PW'])
```

```python
iris.mean()
```

# Checking null values in the dataset

```python
iris.isna().sum()
```

# Visualizations on datasets

# 1.Histogram

```python
plt.hist(data['Classification'],rwidth=0.6)
plt.xlabel('Classification')
plt.ylabel('Frequency')
plt.xticks()
plt.title('Histogram of Classification',fontsize=18)
```

```python
plt.hist(data['SL'],rwidth=0.6)
plt.xlabel('SL')
plt.ylabel('Frequency')
plt.title('Histogram of SL',fontsize=18)
plt.grid()
```

```python
plt.hist(data['SW'],rwidth=0.6)
plt.xlabel('SW')
plt.ylabel('Frequency')
plt.title('Histogram of SW',fontsize=18)
plt.grid()
```

```python
plt.hist(data['PL'],rwidth=0.6)
plt.xlabel('PL')
plt.ylabel('Frequency')
plt.title('Histogram of PL',fontsize=18)
plt.grid()
```

```python
plt.hist(data['PW'],rwidth=0.6)
plt.xlabel('PW')
plt.ylabel('Frequency')
plt.title('Histogram of PW',fontsize=18)
plt.grid()
```
